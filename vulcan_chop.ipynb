I need to parse this into VULCAN banks and got some helpful code to do that from @Simon. He suggested a solution for several banks with uniform size given by `pixels-per-bank=4096` leading to an instrument with a total number of spectra given by `spectrum`. His code was:

```python
pixels_per_bank = 4096
da.coords['bank']= sc.arange(dim='spectrum',start=0, stop = da.sizes['spectrum']) // pixels_per_bank
da.coords['bank']
```

```python
pixels_per_bank = 81920
da.coords['bank'] = sc.arange(dim='spectrum',start=0,stop = da.sizes['spectrum']) // pixels_per_bank
da.coords['bank']
```

Yup...
Next step is to apply this and see how long it takes

```python
time0 = time.time()
pixels_per_bank = 81920
da.coords['bank'] = sc.arange(dim='spectrum',start=0,stop = da.sizes['spectrum']) // pixels_per_bank
da2 = scn.convert(da, 'tof', 'dspacing', scatter=True)
grouped = da2.groupby('bank').bins.concat('spectrum')

runtime = 10813.639 #total run time in s
start = da.bins.coords['pulse_time'].min()
steps = sc.arange(dim='pulse_time', start = 0.0, stop = runtime, step=1.0, unit='s')
steps = sc.to_unit(steps, 'ns').astype('int64')
pulse_times = start + steps

result = sc.bin(grouped, edges=[pulse_times])

dspacing = sc.arange(dim='dspacing', start=0.3, stop = 2.0, step = 0.001, unit='Angstrom')
print('time to complete {}'.format(time.time()-time0))
sc.histogram(result, bins=dspacing).plot()
```
